```python
import cv2
from PIL import Image
import pytesseract

whatsapp_img_path = "C:/Users/Debajyoti/OneDrive/Desktop/Jarves full agent/WhatsAppAssistant/images/test_whatsapp.png"
img = Image.open(whatsapp_img_path)

ocr_data = pytesseract.image_to_data(img,lang="eng",output_type=pytesseract.Output.DICT)
```

```python
print(ocr_data)
```

```python
CHAT_HISTORY_SNIPPET_REGION = (450, 700, 800, 700) 

n_boxes = len(ocr_data["level"])
x_region,y_region,w_region,h_region = CHAT_HISTORY_SNIPPET_REGION
right_alignment_threshold_x = x_region + (w_region * 0.45)
parsed_lines = {}
for i in range(n_boxes):
    if int(ocr_data['conf'][i])>40:
        text = ocr_data['text'][i].strip()
        if not text:
            continue

        block = ocr_data['block_num'][i]
        par = ocr_data['par_num'][i]
        line = ocr_data['line_num'][i]
        left = ocr_data['left'][i] + x_region 

        line_key = (block,par,line)

        if line_key not in parsed_lines:
            parsed_lines[line_key] = {
                'text':[],
                'sender':None,
                'min_left':float('inf'),
                'max_conf':0,
                'full_text':""
            }

        parsed_lines[line_key]['text'].append(text)
        parsed_lines[line_key]['min_left'] = min(parsed_lines[line_key]['min_left'],left)
        parsed_lines[line_key]['max_conf'] = max(parsed_lines[line_key]['max_conf'], int(ocr_data['conf'][i]))

    for key in parsed_lines.keys():
        if parsed_lines[key]['min_left']>right_alignment_threshold_x:
            parsed_lines[key]['sender']="user"
            parsed_lines[key]['full_text'] = " ".join(parsed_lines[key]['text'])
        else: 
            parsed_lines[key]['sender']='Arnab'
            parsed_lines[key]['full_text'] = " ".join(parsed_lines[key]['text'])
            


```

```python
for key,value in parsed_lines.items():
    print(f"{value['sender']}:{value["full_text"]}")
    print(f"Min left: {value["min_left"]}")
```
